```python
import json
from requests import request
from pandas.io.json import json_normalize

def getLatLng(endereco):
    print(endereco)
    try:
        url = "https://maps.googleapis.com/maps/api/geocode/json?sensor=false&key=AIzaSyBvP9gdn_I09X5uPJnkYM4jHIdCdXApogY&address={endereco}".format(endereco = endereco)
        response = request(url=url, method='get')
        geolocal = response.json()
        dado = json_normalize(geolocal['results'])
        return dado['geometry.location.lat'], dado['geometry.location.lng']
    except:
        return 0, 0
```

```python
import time

start_time = time.clock()

redeIndireta['LATITUDE'], redeIndireta['LONGITUDE'] = redeIndireta['END_COMPLETO'].apply(getLatLng)

print(time.clock() - start_time, " seconds")
```
