### DIABETES - Regressão

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfDiabetes = pd.read_csv("diabetes_numeric.csv")
```

```python
dfDiabetes.head(10)
```

```python
dfDiabetes.shape
```

```python
dfDiabetes.describe()
```

```python
dfDiabetes.info()
```

```python
labelDiabetes = np.array(dfDiabetes['c_peptide'])

dfDiabetes = dfDiabetes.drop('c_peptide', axis=1)
dataDiabetes = np.array(dfDiabetes)
```

```python
from sklearn.model_selection import train_test_split

# aplicando a funcao train_test_split para separar os conjuntos de treino e 
# teste segundo uma porcentagem de separação definida. 
train_data, test_data, train_labels, test_labels = train_test_split(dataDiabetes, labelDiabetes, test_size = 0.25, random_state = 42)
```

```python
# importar o modelo de regressão linear
from sklearn.linear_model import LinearRegression 

# treinando o modelo no conjunto de dados de treino
regression = LinearRegression().fit(train_data, train_labels);
```

```python
# aplicando o modelo treinado para a previsão da temperatura 
#em todo o conjunto de teste
predictions1_labels = regression.predict(test_data)

# Exibindo dataframe com valores 10 reais e suas respectivas previsões
p = pd.DataFrame({'Real': test_labels, 'Previsto': predictions1_labels})  
p.head(10)
```

```python
# importar biblioteca para calculo de métricas
from sklearn import metrics  

#avaliando o modelo
print('R2:', metrics.r2_score(test_labels, predictions1_labels)) 
print('MSE:', metrics.mean_squared_error(test_labels, predictions1_labels))  
print('MAE:', metrics.mean_absolute_error(test_labels, predictions1_labels))
```

## BLOODTRANSF - SVC

```python
dfBlood = pd.read_csv("bloodtransf.csv")
```

```python
dfBlood.head(10)
```

```python
dfBlood.shape
```

```python
dfBlood.describe()
```

```python
dfBlood.info()
```

```python
labelBlood = np.array(dfBlood['Class'])

dfBlood = dfBlood.drop('Class', axis=1)
dataBlood = np.array(dfBlood)
```

```python
from sklearn.model_selection import train_test_split

# aplicando a funcao train_test_split para separar os conjuntos de treino e 
# teste segundo uma porcentagem de separação definida. 
train_data, test_data, train_labels, test_labels = train_test_split(dataBlood, labelBlood, test_size = 0.25, random_state = 42)
```

```python
#importar o modelo SVM
from sklearn.svm import SVC

# instanciacao e determinacao dos hiperparâmetros do SVM: tipo de kernel
classifier = SVC(kernel='rbf')

# treinando o SVM
classifier.fit(train_data,train_labels)
```

```python
# aplicando o modelo treinado para a previsão da temperatura 
#em todo o conjunto de teste
predictions2_labels = classifier.predict(test_data)

# Exibindo dataframe com valores 10 reais e suas respectivas previsões
p = pd.DataFrame({'Real': test_labels, 'Previsto': predictions2_labels})  
p.head(10)
```

```python
#avaliando o modelo 
print('Matriz de Confusão\n', metrics.confusion_matrix(test_labels, predictions2_labels)) 
print('\nAcurácia\n', metrics.accuracy_score(test_labels, predictions2_labels)) 
print('\nPrecision\n', metrics.precision_score(test_labels, predictions2_labels)) 
print('\nRecall\n', metrics.recall_score(test_labels, predictions2_labels)) 

print('\nAUCROC\n', metrics.roc_auc_score(test_labels, predictions2_labels))
```

## WINE - KMEANS

```python
dfWine = pd.read_csv("wine.csv")
```

```python
dfWine.head(10)
```

```python
dfWine.shape
```

```python
dfWine['class'].unique()
```

```python
dfWine.describe()
```

```python
dfWine.info()
```

```python
labelWine = np.array(dfWine['class'])

dfWine = dfWine.drop('class', axis=1)
dataWine = np.array(dfWine)
```

```python
from sklearn.model_selection import train_test_split

# aplicando a funcao train_test_split para separar os conjuntos de treino e 
# teste segundo uma porcentagem de separação definida. 
train_data, test_data, train_labels, test_labels = train_test_split(dataWine, labelWine, test_size = 0.25, random_state = 42)
```

```python
# importar o modelo de KMeans
from sklearn.cluster import KMeans

clustering = KMeans(n_clusters = 3, random_state = 42)
 
# treinando o modelo no conjunto de dados de treino
clustering.fit(train_data);
```

```python
# aplicando o modelo treinado para a previsão da temperatura 
#em todo o conjunto de teste
predictions1_labels = clustering.predict(test_data)

# Exibindo dataframe com valores 10 reais e suas respectivas previsões
p = pd.DataFrame({'Real': test_labels, 'Previsto': predictions1_labels})  
p.head(10)
```

```python
#avaliando o modelo

print('Coeficiente de Silhueta\n', metrics.silhouette_score(test_data, predictions1_labels)) 
print('\nDavies-Bouldin Score\n', metrics.davies_bouldin_score(test_data, predictions1_labels)) 

print('\nMatriz de Contingência\n', metrics.cluster.contingency_matrix(test_labels, predictions1_labels)) 
print('\nMutual information\n', metrics.mutual_info_score(test_labels, predictions1_labels)) 
```
